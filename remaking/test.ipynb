```python
from sklearn.model_selection import ParameterSampler
from scipy.stats.distributions import expon, uniform
import numpy as np
rng = np.random.RandomState(0)
param_grid = {'distance':uniform(loc=0.1, scale=0.9), 
              'minimum_samples': [1,3,5,10,20],
              'delta_distance': uniform(loc=0.01, scale=0.09),
              'delta_minimum_samples': [1,2,3],
              'max_iteration': [20,40]
              }
param_list = list(ParameterSampler(param_grid, n_iter=100,
                                   random_state=rng))
rounded_list = [dict((k, round(v, 4)) for (k, v) in d.items())
                for d in param_list]
rounded_list
```

```python
filename = "../data/processed/ATIS.xlsx"
evaluate_dataset = EvaluateDataset(filename=filename, filetype='xlsx', text_column='data', 
                                   target_column='intent', embedder=model.embedding_model)

# results = evaluate_dataset.evaulate_iter_dbscan(parameters)
results = evaluate_dataset.evaulate_iter_dbscan(rounded_list)
result_df = pd.DataFrame.from_dict(results)
result_df

```
